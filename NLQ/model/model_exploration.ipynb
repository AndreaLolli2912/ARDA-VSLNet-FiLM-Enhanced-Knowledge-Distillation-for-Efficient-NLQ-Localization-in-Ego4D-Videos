```python
import torch
from DeepVSLNet import DeepVSLNet
from types import SimpleNamespace

def count_parameters(model: torch.nn.Module) -> int:
    return sum(p.numel() for p in model.parameters())

def get_param_count_for_dim(dim_value: int) -> int:
    # Build a SimpleNamespace so DeepVSLNet.__init__ can access attributes by name
    configs = SimpleNamespace(
        video_feature_dim=1024,
        dim=dim_value,
        drop_rate=0.2,
        num_heads=8,
        max_pos_len=128,
        predictor="bert",
        text_agnostic=False,
    )
    # Instantiate the model and count parameters
    model = DeepVSLNet(configs=configs, word_vectors=None)
    return count_parameters(model)

# Choose a few hidden dims to test:
dim_values = [32, 64, 96, 128]

for d in dim_values:
    total_params = get_param_count_for_dim(d)
    print(f"dim={d} --> {total_params:,} parameters")
```
